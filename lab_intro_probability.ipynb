**Challenge 1**

#### Ironhack Airlines

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats.
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight.

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
from scipy.stats import binom

# Parameters
seats = 450
tickets = 460
show_up_prob = 0.97

# Calculate probability of more than 450 passengers showing up
more_than_seats_probability = binom.cdf(seats, tickets, show_up_prob)

# Subtract from 1 to find the probability for at most 450 passengers showing up
probability = 1 - more_than_seats_probability

print(f"Probability that all passengers get a seat: {probability:.4f}")
print(f"That's approximately {probability * 100:.2f}%")
```

**Challenge 2**

#### Ironhack Call Center

Suppose a customer service representative at a call center is handling customer complaints. Consider the following:
- The probability of successfully resolving a customer complaint on the first attempt is 0.3.


What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

Attempting 2 different ways. The first is Geometric Distribution PMF - it directly calculates the probability of the first success on a specific attempt - "waiting for the first success".


```python
from scipy.stats import geom

# Probability of success
p = 0.3

# At least 3 attempts before success implies failing the first 2.
# Using complementary probability for P(X < 3)
p_at_least_3 = 1 - (geom.pmf(1, p) + geom.pmf(2, p))

print(f"Probability of needing at least three attempts: {p_at_least_3:.4f}")
print(f"That's approximately {p_at_least_3 * 100:.2f}%")
```

This is 2nd Geometric Distribution CDF: Sums probabilities up to a certain attempt, making it useful for calculating the probability of the first success occurring before or at a certain attempt, and it's easily adapted to find the probability of needing more attempts by using its complement.

```python
from scipy.stats import geom

# Probability of success on a single attempt
p = 0.3

# Use CDF to find the probability that the first success occurs on or before the 2nd attempt
p_less_than_3 = geom.cdf(2, p)

# The complement gives the probability that the first success is on or after the 3rd attempt
p_at_least_3 = 1 - p_less_than_3

print(f"Probability of needing at least three attempts: {p_at_least_3:.4f}")
print(f"That's approximately {p_at_least_3 * 100:.2f}%")
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

```python
from scipy.stats import poisson

# Average rate of visits per hour
lambda_visits = 500

# Maximum visits the server can handle
max_visits = 550

# CDF gives the probability of up to 550 visits
prob_up_to_550 = poisson.cdf(max_visits, lambda_visits)

# Complement gives the probability of more than 550 visits
prob_overwhelmed = 1 - prob_up_to_550

print(f"Probability of server being overwhelmed: {prob_overwhelmed:.4f}")
```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
from scipy.stats import poisson

lambda_visits = 500
max_visits = 550

# Probability of not being overwhelmed in a single hour
prob_not_overwhelmed_single_hour = poisson.cdf(max_visits, lambda_visits)

# Probability of not being overwhelmed during any of the 24 hours
prob_not_overwhelmed_full_day = prob_not_overwhelmed_single_hour ** 24

# Probability of being overwhelmed at least once in a day
prob_overwhelmed_at_least_once = 1 - prob_not_overwhelmed_full_day

print(f"Probability of being overwhelmed at least once in a day: {prob_overwhelmed_at_least_once:.4f}")
```

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
from scipy.stats import expon

# The average rate of arrivals (customers per minute)
lambda_rate = 1 / 10  # 0.1 per minute

# Calculate the probability of arrival within the next 5 minutes
prob_within_5_minutes = expon.cdf(5, scale=1/lambda_rate)

print(f"Probability that the next customer will arrive within the next 5 minutes: {prob_within_5_minutes:.4f}")
```

my note: Inverse Relationship: Since λ represents how many times an event happens per unit of time, its inverse (scale) gives the average unit of time per event.

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
# Average arrival rate per minute
lambda_rate = 0.1

# Calculate the probability of no customer arriving within 15 minutes
prob_no_arrival_15_min = 1 - expon.cdf(15, scale=1/lambda_rate)

print(f"Probability of an employee being able to take a break: {prob_no_arrival_15_min:.4f}")
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams.

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
from scipy.stats import norm

# Define the parameters
mu = 150  # mean
sigma = 10  # standard deviation

# Calculate the probability of a weight less than 160 grams
prob_less_than_160 = norm.cdf(160, mu, sigma)

# Calculate the probability of a weight less than 140 grams
prob_less_than_140 = norm.cdf(140, mu, sigma)

# The probability of a weight between 140 and 160 grams
prob_between_140_and_160 = prob_less_than_160 - prob_less_than_140

print(f"Probability that a bird weighs between 140 and 160 grams: {prob_between_140_and_160:.4f}")
```
